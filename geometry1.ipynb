```python
from matplotlib.markers import MarkerStyle
from matplotlib.patches import Ellipse
from matplotlib.patches import Rectangle
from matplotlib.transforms import Affine2D

import math as m

# Create a figure and axis
fig, ax = plt.subplots(subplot_kw={"aspect": "equal"})

a=5.0
b=2.0
c=m.sqrt(a**2+b**2)
squareA = Rectangle(
    xy=(0,0),
    width=a,
    height=a,
    facecolor="none",
    edgecolor="b"
)
squareB = Rectangle(
    xy=(a,a),
    width=b,
    height=b,
    facecolor="none",
    edgecolor="b"
)
squareC= Rectangle(
    xy=(b,0),
    width=c,
    height=c,
    angle=m.degrees(m.atan(b/a)),
    facecolor="none",
    edgecolor="black"
)
squareAB = Rectangle(
    xy=(0,0),
    width=a+b,
    height=a+b,
    facecolor="none",
    edgecolor="r"
)

ax.add_patch(squareAB)
ax.add_patch(squareA)
ax.add_patch(squareB)
ax.add_patch(squareC)

ax.set_xlim(-1,a+b+1)
ax.set_ylim(-1,a+b+1)


# Note: To reverse the orientation arrow, switch the marker type from > to <.

plt.show()
```
